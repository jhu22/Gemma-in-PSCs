```python
def get_xyz_3dH(smiles):
    filename = 'SMILES_'+smiles+'.xyz'
    '''
    #输入smi，获得xyz文件
    '''
    molecule = Chem.MolFromSmiles(smiles)
    molecule = Chem.AddHs(molecule)
    AllChem.EmbedMolecule(molecule, AllChem.ETKDG())
    AllChem.UFFOptimizeMolecule(molecule)
    conf = molecule.GetConformer()
    xyz_coordinates = []
    for atom in molecule.GetAtoms():
        position = conf.GetAtomPosition(atom.GetIdx())
        xyz_coordinates.append((atom.GetSymbol(), position.x, position.y, position.z))
    
    with open(filename, "w") as f:
        # Write header (number of atoms and comment line)
        f.write(f"{len(xyz_coordinates)}\n")
        f.write("Generated by RDKit\n")
        # Write atom data
        for atom, x, y, z in xyz_coordinates:
            f.write(f"{atom} {x:.4f} {y:.4f} {z:.4f}\n")
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
```
